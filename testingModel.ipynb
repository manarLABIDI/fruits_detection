```python
import  numpy as np
import  tensorflow as tf
import matplotlib.pyplot as plt 
```

```python
cnn = tf.keras.models.load_model('/content/drive/MyDrive/detection/trained_model.h5')
```

```python
import cv2
image_path = "/content/drive/MyDrive/dataset/test/freshapples/Screen Shot 2018-06-08 at 4.59.44 PM.png"
img = cv2.imread(image_path)
plt.imshow(img)
plt.title("Test Image")
plt.xticks([])
plt.xticks([])
plt.show()
```

Testing Model

```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(20,20))
input_arr = tf.keras.preprocessing.image.img_to_array(image)#converting single image to array veacause of model
input_arr = np.array([ input_arr ])#converting single image  (2dimensions array)
predictions = cnn.predict(input_arr)
```

```python
print(predictions)

```

```python
print(max(predictions[0]))
```

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/dataset/test',

labels="inferred",
label_mode="categorical",
class_names=None,
color_mode="rgb",
batch_size=10,
image_size=(20,20),
shuffle=False,

)
```

```python
test_set.class_names
```

```python
result_index = np.where(predictions[0] == max(predictions[0]))
print(result_index)
```
